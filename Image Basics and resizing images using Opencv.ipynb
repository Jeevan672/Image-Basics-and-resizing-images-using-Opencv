resizing image

```python
fix_img.shape
```

```python
new_img=cv2.resize(fix_img,(3840,2160))
```

```python
plt.imshow(new_img)
```

```python
#resizing image by ratio
w_ratio=0.5
h_ratio=0.5
```

```python
new_img= cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
ni=cv2.resize(fix_img,(0,0),fix_img,0.8,0.2)
```

```python
plt.imshow(ni)
```

Flip an image

```python
new_img=cv2.flip(fix_img,0)
plt.imshow(new_img)
```

```python
new_img=cv2.flip(fix_img,1)
plt.imshow(new_img)
```

```python
new_img=cv2.flip(fix_img,-1)
plt.imshow(new_img)
```
